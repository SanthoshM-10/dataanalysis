```python
import pandas as pd
import matplotlib.pyplot as plt
p = pd.read_csv("class_marks.csv")
p
```

# Reading the class_marks csv dataset

```python
p.info()
```

# Structure of the dataset is observed

```python
p = p.fillna(0)
```

# Modifying nan with zero

```python
p = p.astype("int64")
p
```

# changing float datatype into integer

```python
p = p.replace(28,29)
```

# Replacing marks 28 with 29

```python
p.shape
```

# No.of rows and columns

```python
p["Q1"]=p["Q1aM4"] + p["Q1bM6"]
p["Q2"]=p["Q2aM6"] + p["Q2bM4"]
p["Q3"]=p["Q3aM5"] + p["Q3bM5"]
p["Q4"]=p["Q4aM3"] + p["Q4bM7"]
p["Q5"]=p["Q5M10"]
p["Q6"]=p["Q6aM4"] + p["Q6bM6"]
p.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True)
```

# Combine columns as single named columns

```python
p.head(10)
```

# Displaying first 10 rows

```python
p.tail(10)
```

# Displaying last 10 rows

```python
p["percentage"]=((p["Total"]/60)*100).round()
p
```

# Calculating percentage with reference of Total

```python
def assign_grade(percentage):
    if percentage >= 60:
        return 'A'
    elif percentage >= 50:
        return 'B'
    elif percentage >= 40:
        return 'C'
    elif percentage >= 30:
        return 'D'
    else:
        return 'F'

p['Grade'] = p['percentage'].apply(assign_grade)
p
```

# Assigning the grade

```python
grade_counts = p["Grade"].value_counts()
plt.figure(figsize=(8, 5))
grade_counts.plot(kind="bar", color="skyblue")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
p=pd.read_csv("class_marks.csv")
p["Q1"]=p["Q1aM4"] + p["Q1bM6"]
p["Q2"]=p["Q2aM6"] + p["Q2bM4"]
p["Q3"]=p["Q3aM5"] + p["Q3bM5"]
p["Q4"]=p["Q4aM3"] + p["Q4bM7"]
p["Q5"]=p["Q5M10"]
p["Q6"]=p["Q6aM4"] + p["Q6bM6"]
p.drop(["Q1aM4", "Q1bM6","Q2aM6","Q2bM4" ,"Q3aM5" , "Q3bM5", "Q4aM3" ,"Q4bM7","Q5M10","Q6aM4","Q6bM6"], axis=1, inplace=True)
total_by_q5 = p.groupby('Q1')['Total'].sum()
plt.figure(figsize=(8, 6))
total_by_q5.plot(kind='pie', autopct='%d%%', colors=['brown', 'orange', 'green', 'blue', 'red'])
plt.ylabel('')
plt.title('Total by Q1')
plt.show()
```
