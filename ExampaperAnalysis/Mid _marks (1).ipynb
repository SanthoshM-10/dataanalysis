```python
import scipy.stats as stats
```

```python
from scipy.stats import ttest_1samp

t_statistic, p_value = ttest_1samp(df[df['SECTION'] == 'ALPHA']['DV'], df["DV"].mean())
print(f"T-statistic: {t_statistic}, P-value: {p_value}")
```

```python
t_statistic, p_value = stats.ttest_1samp(df[df['SECTION'] == 'BETA']['DV'],df.DV.mean())
print(t_statistic , p_value)
```

```python
t_statistic, p_value = stats.ttest_1samp(df[df['SECTION'] == 'GAMMA']['DV'],df.DV.mean())
print(t_statistic , p_value)
```

```python
t_statistic, p_value = stats.ttest_1samp(df[df['SECTION'] == 'DELTA']['DV'],df.DV.mean())
print(t_statistic , p_value)
```

```python
def assign_grade(PP):
    if 18 <= PP <= 20:
        return 'Very Good'
    elif 15 <= PP <= 17:
        return 'Good'
    elif 13 <= PP <= 14:
        return 'Average'
    else:
        return 'Poor'

df['PP programming skills'] = df['PP'].apply(assign_grade)
df
```

```python
def assign_grade(DV):
    if 18 <= DV <= 20:
        return 'Very Good'
    elif 15 <= DV <= 17:
        return 'Good'
    elif 13 <= DV <= 14:
        return 'Average'
    else:
        return 'Poor'

df['DV analytical skills'] = df['DV'].apply(assign_grade)
df
```
