# One sample t-test

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
df_cleaned = df.dropna()
sc_beneficiaries = df_cleaned["total_sc"]
st_beneficiaries = df_cleaned["total_st"]
t_stat_ind, p_value_ind = stats.ttest_ind(sc_beneficiaries, st_beneficiaries, equal_var=False)
print(f"Independent T-Test: t-statistic = {t_stat_ind:.4f}, p-value = {p_value_ind:.4f}")
```

## Independent t-test of comparing of SC vs ST Beneficiaries

```python
aadhar_registered = df_cleaned["total_aadhar"]
mobile_registered = df_cleaned["total_mobileno"]
t_stat_paired, p_value_paired = stats.ttest_rel(aadhar_registered, mobile_registered)
print(f"Paired T-Test: t-statistic = {t_stat_paired:.4f}, p-value = {p_value_paired:.4f}")
```

## Paried t-test of comparing for Aadhar vs Mobile registrations

```python
contingency_table = pd.crosstab(df_cleaned["state_name"], df_cleaned["total_aadhar"] > 0)
chi2_stat, p_value_chi2, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test: chi2-statistic = {chi2_stat:.4f}, p-value = {p_value_chi2:.4f}")
```
